$Mapeo Logistico$

El mapeo logistico $x_n+1=rx_n(1-x_n)$, 
Donde r es el parámetro de control. Las repeticiones iteradas generan una secuencia de puntos conocida como órbita donde $\:\:r\in[0,4]\:\mathrm{y}\:x\in[0,1]$
Usando exponente de Lyapunov $\lambda$ que está dado por $\lambda(x_{0})=\lim_{n\to\infty}\frac{1}{n}\sum_{i=0}^{n-1}\log\left|\:f'(x_{i})\right|\:\:\mathrm{con}\:\:x_{i}=f(x_{i-1}).$ para cada orbita que inicia de $x_0$

```julia
function bifurcacion(x,r,n)                    #Función bifurcación
    for i in 1:n                              
        x = r*x*(1-x)                          #Curva logistica
    end
    return x                                  
    end
```

```julia
function Lya(x,n,r)                       #Función Lyapunov
    l=0    
for i in 1:n                                 
    x = r*x*(1-x)                             #Curva logistica
    l = l+log(abs(r*(1-2*x)))                 #Exponente de Lyaponuv
    end
    return l/n                             
    end
```

```julia
A=linspace(0,1,1000)
B=linspace(1,4,500)
list1 =[]
list2 =[]
for i in 1:length(B)
    push!(list1,bifurcacion.(A,B[i],1000))
    push!(list2,ones(length(A))*B[i])
    end
```

```julia
using PyPlot
plot(B,Lya.(0.1,1000,B),"o",markersize=2)
title("Exponente de lyaponuv")
grid()
```

```julia
using PyPlot
plot(list2,list1,"o",markersize=0.05)
title("Bifurcación")
grid()
```
